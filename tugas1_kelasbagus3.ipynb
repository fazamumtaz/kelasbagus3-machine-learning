## Data Preprocessing

### Hapus kolom yang sekiranya kurang relevan

```python
pd.DataFrame({
    'Sum of MissVal': df[['hotel_brand', 'room_area']].isnull().sum(),
    'Percentage of MissVal': df[['hotel_brand', 'room_area']].isnull().sum() / len(df[['hotel_brand', 'room_area']]) * 100
})
```

```python
df_dropped = df.drop(['sitename', 'qts', 'query_time_stamp',
                      'crawl_date', 'pageurl', 'similar_hotel',
                      'uniq_id', 'hotel_description', 'country',
                      'address', 'hotel_brand', 'room_area' ], axis=1)
```

Berikut alesan kenapa columns di atas itu di-drop:

1. `sitename`, `qts`, `query`, `crawl_date`, `pageurl`, `similar_hotel`, `uniq_id`: data-data ini kemungkinan besar adalah data hasil web scraping, jadi ada kolom kaya `crawl_date` dan yang lain. sementara kolom `pageurl`, `similar_hotel` itu menurutku kayanya ngga begitu relevan
2. `hotel_description`, `address`,: ini kaya free text yang kompleks banget, gaada pola nya, jadi kayanya juga useless kalo dikasih ke model.
3. `hotel_brand`, `room_aera`: banyak sekali missing values di data ini, `hotel_brand` dengan persentase **90.2%** missing values dan `room_area` dengan persentase missing values **71.8%**
4. `country`sedangkan untuk country, karean semuanya itu ada di India, jadi datanya ngga terlalu spesial

### Handling Missing Values

```python
missing_values = pd.DataFrame({
    'Sum of MissVal': df_dropped.isnull().sum(),
    'Percentage of MissVal': df_dropped.isnull().sum() / len(df_dropped) * 100
})
missing_values
```

```python
df_clean = df_dropped.copy()

cols_to_unknown = ['additional_info', 'area',
                   'locality', 'point_of_interest',
                  #  'review_count_by_category',
                   'room_facilities', 'hotel_facilities']
cols_to_zero = ['guest_recommendation', 'site_review_count', 'site_review_rating']

for col in cols_to_unknown:
    df_clean[col] = df_clean[col].fillna('Unknown')

for col in cols_to_zero:
  df_clean[col] = df_clean[col].fillna(0)
```

buat column `object` yang NaN, itu diubah jadi Unknown, dan column yang data type nya `int` atau `float`, diubah jadi 0. Semenjak Kak Bagus bilang lebih mending salah input nilai daripada Mengada-ada nilai yang sebenernya gaada, aku jadi takut pake Mean, Median, atau Mode gitu HEHEHE

```python
def fix_site_stay(srs, col):
  rating = [r.split("::")[-1] for r in srs[col].split("|")]
  rating = [float(r) if len(r) > 0 else 0 for r in rating]
  return rating

ratings = pd.DataFrame(data=df_clean.apply(lambda x: fix_site_stay(x, 'site_stay_review_rating'),
                                                           axis='columns').tolist(),
                       columns=['service_quality_rating', 'amenities_rating',
                                'food_and_drinks_rating', 'value_for_money_rating',
                                'location_rating', 'cleanliness_rating'])
```
